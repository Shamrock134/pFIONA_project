## Introduction

### Programmable Flow Injection (pFI): An Introduction
Programmable flow injection (pFI) addresses the limitations of sequential injection (SI) and flow injection (FI) systems. Unlike SI and FI systems, pFI permits variable flow rates, flow reversal, an expanded range of simultaneous reagents, and generates less hazardous waste. When it comes to reagent-based protocols, the versatility of pFI analyzers makes it a flexible and efficient alternative.

At the heart of the pFI system is an eight-port Lab-On-Valve (LOV) combined with two programmable syringe pumps, each with its own holding cell. This dynamic setup allows for multiple configurations, enabling a wide range of reagent-based protocols (Hatta et al, 2021). A single pFI unit can analyze multiple nutrients and trace metals in seawater including nitrite, phosphate, silicate, and iron, adding efficiency to marine laboratories (Hatta, 2019).

The pFI analyzer mixes chemical solutions at predetermined times, ratios, and volumes using programmable software scripts (FLoZF). This makes pFI units not only versatile, but relatively easy to use. It employs a milli-Q "carrier" to push and pull the solutions though the system. A typical analysis sequence involves two pumps simultaneously aspirating and dispensing solutions in opposing directions into holding cells via the LOV. The sample is sequentially mixed with reagents between holding coils and the solution is dispensed to a flow cell attached to a light source. After a set period that allows the solution to reach a steady state, it is read by a spectrophotometer at a predetermined wavelength (Ruzicka, 2018).

Using precise flow programming, combined with modular-based hardware, pFI offers a compact, reproducible, and versatile platform making it practical for both autonomous and lab-based chemical analyses (Ruzicka, 2016). With future possibilities such as marine carbon chemistry analysis, a pFI unit has the potential to make any marine lab a "One-Stop-Shop".

### The programmable Flow Injection Ocean Nutrient Analyzer (pFIONA)

The programmable Flow Injection Ocean Nutrient Analyzer (pFIONA) is the autonomous version of a pFI benchtop unit designed for in situ marine chemical analysis. It consists of the same principal components as a benchtop unit except the sensitive electronics are encased in a robust waterproof shell for deployment in the field. pFIONA can be remotely operated through an internet-based user interface (UI). This allows technicians to monitor systems, perform calibrations, and modify scripts accordingly from a remote location (Lebrec & Grand, 2024).

Reagents and other working fluids are stored in flexible plasma bags; maximizing deployment durations that currently sit at ~30 days. pFIONA enables the automated analysis of seawater nutrients which current remote sensors cannot measure such as phosphate and silicate. It can also potentially serve as a calibration reference for remote sensors, providing lab-quality measurements that would improve the accuracy of oceanographic models and sensor validation (Lebrec & Grand, 2024).

### Measuring Silicate ($SiO_4^{2-}$) using pFI

Programmable flow injection depends on a colorimetric reaction when analyzing silicate $SiO_4^{2-}$ in seawater. A three-step series of chemical reactions react with
$SiO_4^{2-}$ to form a measurable blue colored solution. The three reagents used in the protocol are ammonium molybdate, oxalic acid and ascorbic acid. 

A seawater sample is combined with ammonium molybdate in a holding coil at a ratio of 1:1 to form silicomolybdate acid. This solution is combined with oxalic acid in the opposite holding coil at a ratio of 1:1. This reduces interference from phosphate by breaking down any phosphomolybdic acid, which would otherwise also be detected.

Ascorbic acid is combined with this intermediate solution at a ratio of 1:1 into the original holding coil to reduce the silicomolybdic acid to silicomolybdenum blue (SiMoB). The final solution is transferred to a flow cell connected where it sits to allow equilibrium and is measured spectrophotometrically at 810 nm. The absorbance is calculated using Beer's Law. Each holding coil, not in use, is flushed with Milli-Q carrier in between mixings to prevent contamination. A calibration curve made with known $SiO_4^{2-}$ standards is used to quantify the concentration of the $SiO_4^{2-}$ in the seawater.

### Silicate Analysis Carryover Problem in pFIONA

When determining the concentration of $SiO_4^{2-}$ in seawater, three samples are typically measured spectrophotometrically per sampling cycle. The mean absorbance of the three measurements is used when determining the concentration of $SiO_4^{2-}$. However, the first absorbance (Abs1) was consistently lower than the subsequent measurements (Abs2:Abs3) throughout the range of concentrations of $SiO_4^{2-}$ typically found in Monter Bay. This implied a carryover (CO) effect possibly caused by residual chemicals left over in the system from the previous sampling cycle.

This led to a negative bias affecting the quantification of concentrations of $SiO_4^{2-}$ in the seawater. Until the carryover issue could be solved, pFIONA was directed to take four measurements as a temporary workaround. Only data from absorbance group (Abs2:Abs4) was used for calculations because these three measurements remained consistent to one another. Since pFIONA operates with a limited supply of reagents, an increase of 33% from the additional samples substantially reduces its time on station. 

Solving the carryover issue would be instrumental in extending the deployment time since PFIONA relies on a restricted supply of reagents and an increase in hazardous waste that also must be held in plasma bags. Although the workaround produced accurate measurements, the carryover problem needed to be solved.

### Definition of Terms:

- ΔCO%(Abs1 → Abs2) - Carrover % between Abs1 and Abs2
- ΔCO%(Abs2 → Abs3) - Carrover % between Abs2 and Abs3
- ΔCO%(Abs3 → Abs4) - Carrover % between Abs3 and Abs4
- ΔCO%(Abs1:3 vs Abs2:4) - Carryover % between absorbance group (Abs1:3) and absorbance group (Abs2:4)
- ΔCO%(Abs1:3 vs Abs1:4) - Carryover % between absorbance group (Abs1:3) and absorbance group (Abs1:4)
- ΔCO%(Abs1:4 vs Abs2:4) - Carryover % between absorbance group (Abs1:4) and absorbance group (Abs2:4)
- RSD%(Abs1:3) - RSD of absorbance group (Abs1:3)
- RSD%(Abs1:4) - RSD of absorbance group (Abs1:4)
- RSD%(Abs2:4) - RSD of absorbance group (Abs2:4)

## Methods

### Experimental Design and Treatments Applied
The experiment was conducted using a benchtop pFI to investigate the silicate carryover issue since it reliably simulates pFIONA. An experiment was conducted to simulate the most extreme carryover scenario regarding concentration differences relative to typical concentrations of $SiO_4^{2-}$ found the Monterey Bay. The experiment was designed to evaluate the efficacy of the treatments applied and made it easier to spot improvements and adjust, if needed.

#### Low-to-High Protocol

A "low-to-high" protocol was employed to maximize any signal indicating carryover. Each cycle was led by measuring 4 blank (~0µmol/L) Milli-Q standards designed to simulate an extremely low concentration. This was followed by a set of four measurements with a known $SiO_4^{2-}$ standard (45µmol/L). The extreme change in concentration exaggerated carryover for easier analysis. Only the 45µmol/L measurements will be compared for statistical analyses. This method was chosen over a traditional calibration curve which relies on smaller incremental concentrations of $SiO_4^{2-}$ of 9, 18, 27, 36, and 45 µmol/L. Each treatment was analyzed multiple times to allow reliable statistical and variability assessment.

#### Treatment Conditions

In total, nine treatments were tested, including a baseline “no treatment” condition. Initially only three treatments were chosen to replicate common field rinsing protocols. Based on these results, further strategies were explored including flushing the system with reagents. The last three treaments contained sodium dodecyl sulfate (SDS) known to be effective on carryover issues (Eskelin et al, 2024;Hatta et al, 2019).

All treatments included:
- No Treatment      — Baseline measurement ran as usual for comparison.
- Sample Flush      — Flushing the system with the 45 µmol/L silicate standard to "condition" the internal system before measurement.
- Milli-Q Flush      — Using Milli-Q carrier to flush the system between measurements.
- NaOH              — Flushing with sodium hydroxide between measurements.
- Other             — This category includes miscellaneous treatments using reagents and other variances to flush between samples.
- Ascorbic Acid     — Rinsing with ascorbic acid before measurements.
- SDS MO            — Adding 11.8 mM sodium dodecyl sulfate (SDS) to the molybdate reagent (no flushing between measurements).
- SDS Carrier       — Adding 11.8 mM SDS to the Milli-Q carrier stream (no flushing between measurements).
- SDS MO & Carrier  — Adding 11.8 mM SDS to both the molybdate reagent and carrier (no flushing between measurements).

Previous protocols, and literature supporting SDS’s role in reducing surface adhesion and carryover were used in deciding which SDS concentrations were selected. Each treatment was assessed using calculated carryover percentages, grouped absorbance comparisons, and relative standard deviation (RSD) metrics. These are covered in further detail in the Results section.

- Acidified molybdate, oxalic acid, ascorbic acid, and standard was mixed using the pFI Silicate Assay (Grand, 2023)

### Calcualations Used for Statistical Analysis

#### For Carryover % (CO%) - Absorbance Measurement Difference
##### Carryover % (1:2)  = (abs2 - abs1) / abs1 * 100
##### Carryover % (2:3)  = (abs3 - abs2) / abs2 * 100
##### Carryover % (3:4)  = (abs4 - abs3) / abs3 * 100

#### For Carryover % - Absorbance Group Difference
##### Group Carryover % (1:3) = mean(Abs1, Abs2, Abs3)
##### Group Carryover % (2:4) = mean(Abs2, Abs3, Abs4)
##### Group Carryover % (1:4) = mean(Abs1, Abs2, Abs3, Abs4)

#### For Relative Standard Deviation (RSD)
##### RSD % (1-3) = std(abs1-3) / mean(abs1-3) * 100
##### RSD % (2-4) = std(abs2-4) / mean(abs2-4) * 100
##### RSD % (1-4) = std(abs1-4) / mean(abs1-4) * 100

These calculations will be used during the statistical analysis:
- CO% will be used to determine if carryover between individual measurements existed before the treatments and after treatments.
- Group CO% will be used to determine if carryover between groups of measurements existed before and after treatments.
- Relative Standard Deviation (RSD) % will be used to determine if RSD % is less than 3% after treatments.

## Results

```python
results_1 = pd.DataFrame([j_stats.wilx_test(co_micro45, "No Treatment", "carry_12"),
    j_stats.wilx_test(co_micro45, "No Treatment", "carry_group_13_24"),
    j_stats.wilx_test(co_micro45, "No Treatment", "rsd_group_13_24")])
display(j_stats.left_aligned(results_1))
```

**Figure 5** - The above table shows the results of a paired Wilcoxon signed-rank test for the baseline treatment. 
- **These results indicate that carryover was present before treatments and that RSD was higher than 3%.**
- *H1₀ = The mean of absorbance 1 (Abs1) is not significantly different from Abs2.*
- Significant difference ΔCO%(Abs1 → Abs2) with a p-value = 0.0001 < 0.05). Null Hypothesis #1 is **rejected**
- *H2₀ = The mean of absorbance group Abs1:Abs3 is not significantly different than group Abs2:4.*
- Significant ΔCO%(Abs1:3 vs Abs2:4) with a p-value = 0.0006 < 0.05. Null Hypothesis #2 is **rejected**
- *H3₀: RSD(Abs1:3) ≤ 3%.*
- The RSD(Abs1:3) is 4.03% which is above 3.0%. Null Hypothesis #3 is **rejected**

## Conclusion

- This project successfully identified and resolved a persistent carryover issue present in the programmable Flow Injection Ocean Nutrient Analyzer (pFIONA). This autonomous system is designed for nutrient analysis in marine environments, albeit this project involves silicate analysis. Testing a "baseline" set of known standards confirmed that the first absorbance measurement (Abs1) was consistently lower than the subsequent ones, indicating carryover between sampling cycles. This resulted in a decreased precision, excess reagent consumption, and reduced deployment times. Since pFIONA's mission depends on staying on station as long as possible and being as precision, solving the carryover problem was essential. 

- Using a benchtop pFI system, a series of treatments were performed using a "low to high" approach consisting of a blank and a high concentration silciate standard. Treatments consisted of flushing with milli-Q, NaOH and the standard itself - all with little or no improvement. When flushing with reagents, ascorbic acid showed a reduction in carryover. After discovering that ascrobic acid contained sodium dodecyl sulfate, and reading literature discussing the effectiveness of SDS on carryover (Eskelin et al, 2024;Hatta et al, 2019), 11.8 nM of SDS was added to each component of the last three treatments; SDS in the modybdate, SDS in the milli-Q carrier, and SDS in both the molydbate and miliQ carrier.

- Statistical tests such as Wilcoxon, Kruskal-Wallis, and Ordinary Least Squares regression revealed that all three SDS treatments significantly reduced carryover. When SDS was added to the molbdate and mill-Q carrier, carryover percentages dropped from ~6.7% to ~1.2%, and RSD in(Abs1:3) was significantly reduced to below 2%, making it undistinguishable from (ABS2:4) which also was reduced, showing that SDS improves overall precision. Although the Wilcoxon test showed a significant difference remains between Abs1 and Abs2 using SDS in the moldydate and the milli-Q carrier, OLS regression analyisis and the Kruskal_Wallis test confirm carryover was shown to now be significantly reduced. The "SDS MO & Carrier" treatment also improved calibration sensitivity ilustrated by steeper slopes and tighter residuals in the calibration curve **(Figure 18)**.

- Prior to this solution, a workaround was used: a fourth measurement was taken and only Abs2–4 were used in calculations. This project confirmed that carryover did not propagate past the first absorbance measurement supporting evidence that the previous workaround resulted in accurate and precise sampling. Reagent consumption and waste has been effectively reduced by 25% by eliminating the need for a workaround. Since the success of pFIONA is measured by precision and endurance, these results directly support this mission.

- While the most pronounced carryover occurred between Abs1 and Abs2, the grouped absorbance comparison (Abs1–3 vs Abs2–4) was the best overall indicator based on its system performance performed during the workaround. The Wilcoxon test showing no significant difference between these two groups using SDS, and carryover dropped from 2.13% to 0.22% (representing a relative drop of 85%). RSD was also reduced by 64%, from 4.03% to 1.46%.y 64% from 4.03% to 1.46%.

- In summary, this study not only confirms that there was a carryover problem, but that is has effectively been resolved through progressive treatment strategies. Most notably was the introduction of SDS into the system that showed the best results improving analytical performance.  These findings improve the efficiency of pFIONA, making her more reliable, thereby improving overall autonomy, setting the stage for further possibilities such as the analysis of pH and total alkalinity.

- Concluding this study, SDS should be tested in pFIONA under read-world conditions either in a shore station or on board a research vessel.
  
**Apply SDS to other nutrient analysis**

- SDS has shown to have a positve effect on the silcate analysis and adding it to other nutirent and possibly trace metal assays should be explored.

**Expand pFI to include marine carbon chemistry**

- Now that SDS has proven to have a significant effect on precision, more sensitve analysis such as pH and total alkalinity should be the next step for pFI.
- With the possibility of carbon chemisty analysis, this would make a pFI unit a "One stop shop" for marine chemical analysis.
  
**Looking into improving long-term reagent stability**

- The success of pFIONA depends on the abiliity to stay on station for extended periods. This will require optimizing the shelf-life of a reagent which is already something that needs to be pursued (Lebrec & Grand, 2024)
